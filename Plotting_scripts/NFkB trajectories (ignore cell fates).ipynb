```julia
function processAllCell1!(allCells, cellMatrixA, cellMatrixC)
    for i in 1:lastindex(allCells)
        push!(cellMatrixA, allCells[i].NuclearRelA);
        push!(cellMatrixC, allCells[i].NuclearcRel);
    end
end
```

```julia
function processAllCell2!(allCells, cellMatrixA, cellMatrixC)
    for i in 1:lastindex(allCells)
        push!(cellMatrixA, allCells[i].NFkBsolution[NA50, :]+allCells[i].NFkBsolution[NA52, :]);
        push!(cellMatrixC, allCells[i].NFkBsolution[NC50, :]+allCells[i].NFkBsolution[NC52, :]);
    end
end
```
